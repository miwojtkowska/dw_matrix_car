```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg(np.median)
 .sort_values(ascending = False)
 .head(50)
).plot(kind='bar', figsize=(15,5))
```

```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.mean, np.median, np.size])
 .sort_values(by='mean', ascending = False)
 .head(50)
).plot(kind='bar', figsize=(15,15),subplots=True)
```

```
def group_plot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort = 'mean', top=50, subplots=True):
  return(
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(by=feat_sort, ascending = False)
    .head(top)
  ).plot(kind='bar', figsize=(15,8),subplots=subplots)
```

```
group_plot('param_marka-pojazdu');
```

```
group_plot('param_country-of-origin');
```

```
group_plot('param_kraj-pochodzenia', feat_sort = 'size');
```

```
group_plot('param_kolor', feat_sort = 'mean');
```
